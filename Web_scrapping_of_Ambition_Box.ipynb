```python
Company = soup.find_all('div', class_='company-content-wrapper')
```

```python
len(Company)
```

```python
name = []
rating = []
reviews = []
ctype = []
hq = []
old = []
employees = []
des = []

for i in Company:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('p', class_='rating').text.strip())
  reviews.append(i.find('a', class_='review-count').text.strip())
  ctype.append(i.find_all('p', class_= 'infoEntity')[0].text.strip())
  hq.append(i.find_all('p', class_= 'infoEntity')[1].text.strip())
  old.append(i.find_all('p', class_= 'infoEntity')[2].text.strip())
  des.append(i.find('p', class_='description').text.strip())
  try:
    employees.append(i.find_all('p', class_= 'infoEntity')[3].text.strip())
  except:
    employees.append(np.nan)

 
df = {'name': name, 'rating' : rating, 'reviews' : reviews, 'Company_Type' : ctype, 'Hq' : hq, 'old' : old, 'employees' : employees, "description" : des} 
data = pd.DataFrame(df)
  
```

```python
final = pd.DataFrame()

for j in range(1, 300):
  url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
   
  headers = {'User-Agent':'Mozilla/5.0 (Window s NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage = requests.get(url,headers = headers).text
  soup = BeautifulSoup(webpage, 'lxml')
  Company = soup.find_all('div', class_='company-content-wrapper')

  name = []
  rating = []
  reviews = []
  ctype = []
  hq = []
  old = []
  employees = []
  des = []

  for i in Company:
    try:
      name.append(i.find('h2').text.strip())
    except:
      name.append(np.nan)

    try:
      rating.append(i.find('p', class_='rating').text.strip())
    except:
      rating.append(np.nan)

    try:
      reviews.append(i.find('a', class_='review-count').text.strip())
    except:
      reviews.append(np.nan)

    try:
      ctype.append(i.find_all('p', class_= 'infoEntity')[0].text.strip())
    except:
      ctype.append(np.nan)

    try:
      hq.append(i.find_all('p', class_= 'infoEntity')[1].text.strip())
    except:
      hq.append(np.nan)

    try:
      old.append(i.find_all('p', class_= 'infoEntity')[2].text.strip())
    except:
      old.append(np.nan)

    try:
      des.append(i.find('p', class_='description').text.strip())
    except:
       des.append(np.nan)

    try:
      employees.append(i.find_all('p', class_= 'infoEntity')[3].text.strip())
    except:
      employees.append(np.nan)

 
  df = {'name': name, 'rating' : rating, 'reviews' : reviews, 'Company_Type' : ctype, 'Hq' : hq, 'old' : old, 'employees' : employees, "description" : des} 
  data = pd.DataFrame(df)

  final = final.append(data, ignore_index=True)

```
